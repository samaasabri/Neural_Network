```python
Input_layer=[1,1,1,1,1]
hidden_layers=3
neurons=[10,10,10]
activision=1
Target=[0,0,1]
```

```python
import numpy as np

#Both parameters taken from GUI
#ex >>(3,[10,10,10])

def Intialize_weights(hidden_layers,neurons):
  wieght_matrices=[]
  for i in range(hidden_layers+1):
    w_i=[]
    if i==0:
      w_i=np.random.rand(neurons[i],5)
    elif i== hidden_layers:
      w_i=np.random.rand(3,neurons[i-1])
    else:
      w_i=np.random.rand(neurons[i],neurons[i-1])


    #print(w_i.shape)
    #print('----------------------------------------')
    wieght_matrices.append(w_i)

  return  wieght_matrices

wieght_matrices=Intialize_weights(hidden_layers,neurons)
```

# **Forward**

```python
def feedForward(Input_layer,neurons,wieght_matrices,activision):
    #activison = 1 ---> sigmoid
    #activison = 0 ---> tengent
  neurons.append(3)
  #contains all f for each layer
  array_of_fs=[]

  def sigmoid(f_i):
    return 1/(1+np.exp(-f_i))

  def tengent(f_i):
    return np.tanh(f_i)

  for i in range(hidden_layers+1):
    #print("Layer: ",i)

    #f contains net value for wach neuron
    f=[]

    for n in range(neurons[i]):
      #print("Neuron: ",n)
      #print(wieght_matrices[i][n])

      if i==0:
        f_i=np.dot(wieght_matrices[i][n],Input_layer)
        if activision:
            f_i = sigmoid(f_i)
        else:
            f_i = tengent(f_i)
        f.append(f_i)
      else:
        f_i=np.dot(wieght_matrices[i][n],array_of_fs[i-1])
        if activision:
            f_i = sigmoid(f_i)
        else:
            f_i = tengent(f_i)
        f.append(f_i)


    array_of_fs.append(f)
  return  array_of_fs
```

```python
wieght_matrices=Intialize_weights(hidden_layers,neurons)
array_of_fs=feedForward(Input_layer,neurons,wieght_matrices,0)
print(array_of_fs)
print(wieght_matrices)
```

# **BackPropagation**

```python
def calc_weight_arr(arr_of_arrs,idx):
  arr_of_wights=[]
  for i in arr_of_arrs:
    arr_of_wights.append(i[idx])
  return arr_of_wights
```

```python
def sigmoid_dash(f_i):
    return ((f_i)*(1-f_i))

def tengent_dash(f_i):
    return (1-pow(np.tanh(f_i),2))

```

```python
def backPropagate(Target,neurons,wieght_matrices,activision,array_of_fs):

  array_of_errors=[]
  j=-1

  for i in range(hidden_layers,-1,-1):
    print("Layer:" ,i)
    error=[]

    for n in range(neurons[i]):
      print("Neuron: ",n)
    #check activision function
      if activision:
        dash = sigmoid_dash(array_of_fs[i][n])

        if i==hidden_layers:
          print('Target:  ',Target[n])
          error_i=(Target[n]- array_of_fs[i][n])*dash

        else:
          #print("Array of errors: ",array_of_errors[j])
          # print("Weight matrix: ",wieght_matrices[i][n])
          wights=calc_weight_arr(wieght_matrices[i+1],n)
          # print("Weight matrix:",wights)
          # print("N is : ",n)

          error_i= np.dot(wights,array_of_errors[j]) *dash
         #continue


      else:
          error_i = tengent_dash(array_of_fs[i][n])

      error.append(error_i)
      #print(error)
      #print(array_of_fs[i][n])

    array_of_errors.append(error)
    print("Array of errors: ",array_of_errors)
    print("----------------")
    j+=1
  return array_of_errors
```

```python
array_of_errors=backPropagate(Target,neurons,wieght_matrices,activision,array_of_fs)
```

#**Update weights**

```python
array_of_errors.reverse()
print(array_of_errors)
```

```python
def Update_wieghts(array_of_errors,wieght_matrices,Input_layer,learning_rate,hidden_layers,neurons,array_of_fs):

  for i in range(hidden_layers+1):
    #print("Layer: ",i)


    for n in range(neurons[i]):
      #print("Neuron: ",n)
      #print(wieght_matrices[i][n])
      for j in range(len(wieght_matrices[i][n])):


         if i==0:
           wieght_matrices[i][n][j]=wieght_matrices[i][n][j]+learning_rate*array_of_errors[i][n]*Input_layer[j]

         else:
           wieght_matrices[i][n][j]=wieght_matrices[i][n][j]+learning_rate*array_of_errors[i][n]*array_of_fs[i][n]


  return wieght_matrices


```

```python
updated_weights=Update_wieghts(array_of_errors,wieght_matrices,Input_layer,0.1,hidden_layers,neurons,array_of_fs)
updated_weights
```

```python
def train(input,target,hidden_layers_num,neurons_num,Activation_used,learninR,num_epochs):
  net=Intialize_weights(hidden_layers=hidden_layers_num,neurons=neurons_num)
  for i in range(num_epochs):
    for i in range(len(input)):
      FF=feedForward(input[i],neurons_num,net,Activation_used)
      array_of_error=backPropagate(target[i],neurons_num,net,Activation_used,FF)
      net=Update_wieghts(array_of_error,net,input[i],learninR,hidden_layers_num,neurons_num,FF)

  return net


```

```python
def predict(input,neurons_num,net,Activation_used):
  predictions=[]
  for sample in input:
    FF=feedForward(sample,neurons_num,net,Activation_used)
    predictions.append(FF.index(max(FF[-1])))

  return predictions
```
